# 1.Import Naccssory Liabrary 


```python
# 1.Import Naccssory Liabrary 
import skimage as sk
import cv2
import numpy as np
import matplotlib.pyplot as plt
import imutils
```

```python
def show_image(img):
    """
    This Method to disply image in 10*10 scale
    """
    fig = plt.figure(figsize=(10,10))
    ax  = fig.add_subplot(1,1,1, xticks=[], yticks=[])
    ax.imshow(img, cmap='gray')
    
```

# 2. Display image

```python
img1 = cv2.imread('image1(1).jpg', 0)
show_image(img1)
```

 # 3. Preporccssing

```python
# Blur the image 
image_blur = cv2.medianBlur(img1,15)
show_image(image_blur)
```

```python
threshed = cv2.Canny(image_blur,128,128)
show_image(threshed)
```

```python
dilate = cv2.dilate(threshed, (1,1), 2)
```

# 4. Find contours

```python
contours, hirarchy = cv2.findContours(dilate.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(img1, contours, -1, (0, 255, 0), 2)


show_image(img1)
```

# 5. Disply number of coins

```python
# create dictionry to map interval for each coin area
dict_for_weghits = {
    1: [15, 25], # if it is in this range so it is in 1 
    
    5: [50,59],# if it is in this range so it is in 5
    
    10:[60,70]# if it is in this range so it is in 10
}
```

```python
def display(img,count,img_path,cmap="gray"):
    weights = []
    
    for (i,c) in enumerate(contours):
        # compute the center of the contour
        M = cv2.moments(c)
        weight = cv2.contourArea(c )/100
        if dict_for_weghits[1][0]<= weight <= dict_for_weghits[1][1]:
            weights.append(1) 
        elif dict_for_weghits[5][0]<= weight <= dict_for_weghits[5][1]:
            weights.append(5) 
        else:
            weights.append(10) 
        cX = int(M["m10"] / M["m00"])
        cY = int(M["m01"] / M["m00"])
        cv2.putText(img, "#{}".format(i+1), (cX - 20, cY - 20),
            cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
        
    f_image = cv2.imread(img_path)
    f, axs = plt.subplots(1,2,figsize=(20,20))
    axs[0].imshow(f_image,cmap="gray")
    axs[1].imshow(img,cmap="gray")
    axs[1].set_title("Total Money Count = {} \n Total weights is {}".format(count, sum(weights)))
    
```

```python
display(img1,len(contours), "image1(1).jpg")
```
